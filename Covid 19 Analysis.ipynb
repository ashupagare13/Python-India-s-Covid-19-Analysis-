```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pylab as plt
import warnings
import plotly.express as px
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
vaccination = pd.read_csv('D:/Data Analytics/Projects/Covid 19/COVID-19 India Statewise Vaccine Data.csv')
cases = pd.read_csv('D:\Data Analytics\Projects\Covid 19\Latest Covid-19 India Status.csv')
```

```python
vaccination.head(10)
```

```python
vaccination.rename(columns = {'Population':'Total Vaccinated People'}, inplace = True)
vaccination = vaccination.replace('Dadra and Nagar Haveli and Daman and Diu', 'Daman and Diu')
vaccination.head(10)
```

```python
cases.head(10)
```

```python
cases.rename(columns = {'Population':'Total State/UTs Population'}, inplace = True)
cases=cases.replace('Dadra and Nagar Haveli and Daman and Diu', 'Daman and Diu')
cases.head(10)
```

# Total State/UTs Population and Cases

```python
#create the base axis to add the bars to
fig, ax = plt.subplots(1,1, figsize = (20,8))
#extract the labels
label = data["State/UTs"]
#use this to create x ticks to add the data to
x = np.arange(len(label))
#set a width for each bar 
width = 0.3
#create out first bar
#x will be the centre of the bars
#so set the centre of the first bar to be 1/2 width away to the left
rect1 = ax.bar(x - width/2,
              data['Total State/UTs Population'],
              width = width,
               color='red',
               label = 'Total State/UTs Population')
#create the second bar
#with a centre of the second bar to be 1/2 width away to the right
rect2 = ax.bar(x + width/2,
              data['Total Cases'],
              width = width,
               color='darkred',
              label = 'Total Cases')
#add the labels to the axis
ax.set_ylabel("Total Population",
             fontsize = 20,
             labelpad = 20)
ax.set_xlabel("State/UTs",
             fontsize = 20,
             labelpad =20)
ax.set_title("Total State/UTs Population and Cases",
             fontname="Times New Roman",
            fontsize = 30,
            pad = 20)
#set the ticks
ax.set_xticks(x)
ax.set_xticklabels(label)
#add the legend
#using the labels of the bars
ax.legend(title = "Distribution",
         fontsize = 16,
         title_fontsize = 20)
#x-axis
ax.tick_params(axis = "x", which = 'both', labelsize=15, rotation=90)
#y-axis
ax.tick_params(axis = "y", which = 'both', labelsize=15)
```
